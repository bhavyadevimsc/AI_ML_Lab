```python
data = pd.read_csv('income.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.show()
```

```python
m = 0
c = 0

L = 0.0001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    
print (m, c)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=np.linspace(-10,10,500)
Y=X**2-25
plt.plot(X,Y)
```

```python
# Initial values of the coefficients
beta0 = 0
beta1 = 0
beta2 = 0

L = 0.0001  # Learning rate
epochs = 1000  # Number of iterations

n = float(len(X))  # Number of data points

# Performing Gradient Descent
for i in range(epochs):
    # Predicted value of Y
    Y_pred = beta0 + beta1 * X + beta2 * X**2
    
    # Compute derivatives (gradients)
    D_beta0 = (-2 / n) * sum(Y - Y_pred)  # Derivative wrt beta0
    D_beta1 = (-2 / n) * sum(X * (Y - Y_pred))  # Derivative wrt beta1
    D_beta2 = (-2 / n) * sum((X**2) * (Y - Y_pred))  # Derivative wrt beta2
    
    # Update coefficients
    beta0 = beta0 - L * D_beta0
    beta1 = beta1 - L * D_beta1
    beta2 = beta2 - L * D_beta2

# Final coefficients
print(beta0, beta1, beta2)
```

```python
# Making predictions
Y_pred = beta0 + beta1 * X + beta2 * X**2

plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```
