```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#step 1: create some sample data
data = np.random.rand(100,10)
np.set_printoptions(precision=4, suppress=True)
print("Formated Array:")
print(data)
```

```python
#step 2: standerdize the data
mean= np.mean(data,axis=0)
std_dev=np.std(data,axis=0)
std_data=(data-mean)/std_dev
mean1=np.mean(std_data,axis=0)
std_dev1=np.std(std_data,axis=0)
print(mean,std_dev)
print(mean1,std_dev1)
print(std_data)
```

```python
cov_matrix=np.cov(std_data,rowvar=False)
size_cc=cov_matrix.size
shape_cc=cov_matrix.shape
print(size_cc,shape_cc)
print(cov_matrix)
```

```python
eigenvalues,eigenvectors=np.linalg.eigh(cov_matrix)
print(eigenvalues)
print(eigenvectors)
```

```python
sorted_indices=np.argsort(eigenvalues)[::-1]
eigenvalues=eigenvalues[sorted_indices]
eigenvectors=eigenvectors[:,sorted_indices]
print(eigenvalues)
print(eigenvectors)
```

```python
n_components=5
```

```python
selected_eigenvectors=eigenvectors[:,:n_components]
print(selected_eigenvectors)
```

```python
final_result=np.dot(std_data,selected_eigenvectors)
```

```python
print("Final result after PCA:")
print(final_result)
```

```python
if n_components==2:
    plt.scatter(final_result[:,0],final_result[:,1])
    plt.xlabel('Prinicpal Component 1')
    plt.ylabel('Prinicpal Component 2')
    plt.title("PCA Result")
    plt.show()
```

```python
#step11: Compute reconstruction error(Loss)
reconstructed_data=np.dot(final_result,selected_eigenvectors.T)

#transform back to original scale
reconstructed_data_original_scale=reconstructed_data * std_dev + mean
#Calculate Mean Squared Error b/w original and reconstructed data       
mse=np.mean((data - reconstructed_data_original_scale)**2)
print("Reconstructed Loss:", mse)
```
